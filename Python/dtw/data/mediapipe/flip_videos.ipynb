# Flip videos

```python
### load in required packages
import os
from pathlib import Path #for setting paths
# from moviepy import VideoFileClip, vfx
import ffmpeg
from tqdm import tqdm #for progress bar


### define folders
cwd = Path(os.getcwd())
root_f = cwd.parent.parent.parent.parent.absolute() #get the path to three folders above the current folder (root folder)
media_f = os.path.join(root_f, "05_data", "01_interaction", "02_media")
# list up all the folders that doesn't have "_" and "p" in the folder name
media_folders = [f for f in os.listdir(media_f) if "_" not in f and "p" not in f]
print(media_folders)

outputf_flipped = "flipped_videos/"
```

```python
for foldername in media_folders:
    processed_folder = os.path.join(media_f, foldername, "processed")

    if any(fn.startswith(foldername) for fn in os.listdir(outputf_flipped)):
        continue # skip if the file already exists

    for filename in tqdm(os.listdir(processed_folder)):
        if filename.endswith(".mp4"):
            vfile = os.path.join(processed_folder, filename)
            output_file = outputf_flipped + filename
            ### flip the video
            input_video = ffmpeg.input(vfile)
            flipped_video = input_video.hflip()
            audio = input_video.audio
            out = ffmpeg.output(flipped_video, audio, output_file)
            out.run()
```
